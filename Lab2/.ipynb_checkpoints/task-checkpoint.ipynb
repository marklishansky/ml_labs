# 2. Описать кратко постановку задачи, что от вас хотят. Какие есть переменные. Целевое событие непрерывно (предсказываем число от -inf до +inf) 
либо дискретно (предсказываем класс из конечного множества вариантов, например 0 или 1, или какое-то число в диапазоне [0; 10])

Разработать модель для предсказания целевого значения на основе предоставленного датасета. Целевое событие — это числовое значение 
(непрерывная переменная), которое нужно предсказать. Задача заключается в том, чтобы использовать имеющиеся данные для построения модели, 
которая сможет точно предсказывать это значение для новых данных.

Переменные в датасете:

age: возраст застрахованного лица (целое число).
sex: пол застрахованного лица (строка, возможно "male" или "female").
bmi: индекс массы тела (число с плавающей точкой).
children: количество детей, на иждивении у застрахованного (целое число).
smoker: статус курения (строка, возможно "yes" или "no").
region: регион проживания (строка, возможно обозначения как "northeast", "northwest", "southeast", "southwest").
charges: медицинские расходы (число с плавающей точкой) — это целевая переменная, которую нужно предсказать.

Целевая переменная:

charges: это непрерывная переменная, представляющая собой сумму медицинских расходов, которую нужно предсказать.

Тип задачи:
Задача является задачей регрессии, так как требуется предсказать непрерывное числовое значение (медицинские расходы).


# 3. Построить распределение целевой переменной в виде гистограммы, сделать промежуточные выводы (обратите внимание на однородоность распределения и возможный дисбаланс). Посчитайте количество уникальных значений целевой переменной.

```python
import pandas as pd
import matplotlib.pyplot as plt

# загрузка данных
path_to_dataset = '/Users/daniil/nn_ml_practice/insurance.csv'
data = pd.read_csv(path_to_dataset, sep=',')

# построение гистограммы целевой переменной
plt.figure(figsize=(10, 6))
plt.hist(data['charges'], bins=50, edgecolor='k')
plt.title('Распределение медицинских расходов (charges)')
plt.xlabel('Медицинские расходы (charges)')
plt.ylabel('Частота')
plt.grid(True)
plt.show()

# количество уникальных значений целевой переменной
unique_values_count = data['charges'].nunique()
print(f'Количество уникальных значений в столбце "charges": {unique_values_count}')
```

Промежуточные выводы:

Распределение медицинских расходов неоднородно и смещено вправо, что говорит о том, что большинство наблюдений имеют низкие значения расходов, а с увеличением суммы расходов количество наблюдений уменьшается.

Есть заметный дисбаланс в данных по медицинским расходам, с большинством наблюдений с низкими расходами. Количество уникальных значений в столбце "charges": 1337.

# 7. Постройте гистограммы 5 любых признаков (из множества X, или как оно изначально у нас называется data.data). Если видите какое-то смещение, несимметричность и прочее, опишите это словами в ноутбуке.

```python
import matplotlib.pyplot as plt

features = ['age', 'bmi', 'children', 'smoker', 'region']

for feature in features:
    plt.figure(figsize=(8, 6))
    plt.hist(data[feature], bins=30, edgecolor='k')
    plt.title(f'Histogram of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

Возраст: Много людей в датасете моложе 20 лет. Однако есть необычное чередование столбцов частоты на уровнях примерно 25 и 55, что может указывать на неоднородность распределения. Возможно, это связано с особенностями выборки данных.

BMI: Наибольшее количество людей имеют средний уровень индекса массы тела (в районе 30), что может указывать на то, что большинство людей в датасете склонны к избыточному весу. Распределение имеет разброс данных с уменьшением и увеличением индекса отностильно среднего.

Дети: Количество детей у большинства людей в датасете невелико, и с увеличением количества детей уровень частоты равномерно уменьшается.

Курение: Большинство людей в датасете не курят, что подтверждается высоким уровнем частоты для некурящих (больше 1000), в то время как количество курящих людей гораздо меньше (около 300). Это указывает на дисбаланс в данных по курению.

Регион: Гистограмма показывает, что количество людей из разных регионов примерно одинаково, так как уровень частоты примерно одинаков для всех регионов.

# 8. Постройте графики зависимости 5 любых (на ваш выбор) переменных от целевой переменной (если переменных меньше, чем 5, то сделайте столько, сколько получится). Сделайте вывод, можно ли использовать эти переменные для прогнозирования целевой переменной (иначе говоря, есть ли какая-то взаимосвязь между y-переменной и X-переменной)

```python
import matplotlib.pyplot as plt

features = ['age', 'bmi', 'children', 'smoker', 'region']

for feature in features:
    plt.figure(figsize=(8, 6))
    data.boxplot(column='charges', by=feature)
    plt.title(f'Зависимость между {feature} и целевой переменной (charges)')
    plt.xlabel(feature)
    plt.ylabel('charges')
    plt.grid(True)
    plt.show()
```

Вывод по графикам:
Для прогнозирования целевой переменной можно использовать переменные age, smoker и bmi. Возможно можно использовать children, но недостаточный размер выборки по людям с большим количеством детей. На графике зависимости переменной возраста от целевой видно, что размер медицинских затрат равномерно возрастает с увеличением возраста. На графике зависимости индекса массы тела от целевой заметно, что люди с индексом выше 30 имеют повышенный счет на медицину. На графике зависимости переменной курения от целевой очевидна разница между более высоким уровнем у курильщиков, чем у некурящих.
